```python
Q1 = df['adr'].quantile(0.25)
Q3 = df['adr'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers=df[(df['adr']< lower_bound)|(df['adr']> upper_bound )]
```

```python
print(f"ADR outliers: {len(outliers)}")
print(f'lower bound: {lower_bound}')
print(f'upper bound: {upper_bound}')
```

```python
q1=df['lead_time'].quantile(0.25)
q3=df['lead_time'].quantile(0.75)
lower=q1-1.5*(q3-q1)
upper=q3+1.5*(q3-q1)
outliers_2=df[(df['lead_time']< lower )|(df['lead_time']>upper)]
```

```python
print(f"Lead Time Outliers: {len(outliers_2)}")
print(f'lower bound: {lower}')
print(f'upper bound: {upper}')
```

##Data Quality Issues

1. **Outliers**
   - `adr`: 3,793 entries exceed the IQR and labelled as outliers
   - `lead_time`: 3,005 entries flagged as outliers.
   - These may distort statistical analysis and visualizations and show huge number of variation in data.

2. **Missing Values**
   - `company`: 112,593 missing
   - `agent`: 16,340 missing
   - `country`: 488 missing
   - `children`: 4 missing
   - I suggest the column ['country'] to be dropped as it is mildly relevant to our future model detection.

3. **Duplicates**
   - Found **31,994 duplicated rows** using `df.duplicated().sum()`
   - These may inflate counts, bias averages, and misrepresent booking behavior.


#Phase 2

Company:

Out of 119390 entries, 112,593 are missing and 6797 are not missing thus we can replace the null values with Zero.

```python
df['company'].fillna(0, inplace=True)
```

Country:

Out of 119390 entries, 488 are missing thus instead of filling them with incorrect submissions we will assign them as 'Unknown' to be treated as its own type

```python
df['country'].fillna('Unknown', inplace=True)
```

Children:

Out of 119390 entries, only 4 is missing thus its safe to fill in with the mode value

```python
mode_children=df['children'].mode()[0]
df['children'].fillna(mode_children, inplace=True)
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['adr']=df['adr'].apply(lambda x: min(x,1000))
```

The upper bound for 'adr' column is 211.065 yet we have values crossing the 5000 mark making our boxplot inaccurate which is why i used '.apply' to apply a lambda function that takes the value 'x' of each entry and compares it to '1000' and choses the minimum value between them to set it as the official entry.

```python
df['lead_time']=df['lead_time'].apply( lambda x:min(x,500))
```

The upper bound for 'lead_time' column is 373 yet we have values crossing the 700 mark making our boxplot inaccurate which is why i used '.apply' to apply a lambda function that takes the value 'x' of each entry and compares it to '500' and choses the minimum value between them to set it as the official entry.
